```python
import numpy as np
import os

def print_particle_type_details(data_path, file_name):
    # Use np.memmap to read the file
    particle_type = np.memmap(os.path.join(data_path, file_name), dtype=np.int64, mode="r")

    # Print the details of the particle_type array
    print("particle_type shape:", particle_type.shape)
    print("particle_type length:", len(particle_type))
    print("particle_type dtype:", particle_type.dtype)

    # If the array is structured, print the column names
    if hasattr(particle_type, 'dtype') and particle_type.dtype.names is not None:
        print("Available columns (fields):", particle_type.dtype.names)
    else:
        print("The dataset does not contain structured columns.")

    # Set NumPy print options to avoid truncation
    np.set_printoptions(threshold=np.inf, linewidth=np.inf)
    
    
    # Set NumPy print options to avoid truncation
    #np.set_printoptions(threshold=np.inf, linewidth=np.inf)

    # Print the first 100,000 entries of the particle_type array
    num_to_print = min(1000, len(particle_type))  # Ensure we don't exceed the length
    print(f"First {num_to_print} entries of particle_type:")
    print(particle_type[:num_to_print])    

    # Print the full contents of the particle_type array
    #print("Full contents of particle_type:")
    #print(particle_type)
```

```python
###############
## SHOW JSON ##
###############
import json

def read_and_print_json(file_path):
    """Reads a JSON file and prints its contents in a pretty format."""
    try:
        with open(file_path, 'r') as file:
            metadata = json.load(file)
        print(json.dumps(metadata, indent=4))  # Pretty print with an indent of 4 spaces
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
    except json.JSONDecodeError:
        print(f"Error: The file '{file_path}' is not a valid JSON file.")
    except Exception as e:
        print(f"An error occurred: {e}")
```
